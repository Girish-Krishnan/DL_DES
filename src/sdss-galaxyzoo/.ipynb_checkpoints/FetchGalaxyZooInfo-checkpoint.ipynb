```python
def hms2deg(rms_str):
    neg = (rms_str[0] == '-')
    if rms_str[0] in ('-', '+'):
        rms_str = rms_str[1:]
    h, m, s = rms_str.split(':')
    hs = int(h) * 3600
    ms = int(m) * 60
    ts = hs + ms + float(s)
    frac = ts / 86400
    deg = 360 * frac
    return - deg if neg else deg

def dms2deg(rms_str):
    neg = (rms_str[0] == '-')
    if rms_str[0] in ('-', '+'):
        rms_str = rms_str[1:]
    d, m, s = rms_str.split(':')
    ds = int(d)
    ms = int(m) * 60
    deg = ds + (ms + float(s)) / 3600
    return - deg if neg else deg
```

```python
gz_ds['RA_deg'] = [hms2deg(x) for x in gz_ds['RA']]
gz_ds['DEC_deg'] = [dms2deg(x) for x in gz_ds['DEC']]
```

```python
plt.hist2d(gz_ds['RA_deg'], gz_ds['DEC_deg'], bins=[200, 100], range=[[0, 360], [-90, 90]], cmap='hot_r')
plt.colorbar(label='Count')
plt.xlabel('RA')
plt.ylabel('Dec')
```

```python
des_range = [
    # [[ra_min, ra_max], [dec_min, dec_max]]
    [[0, 45.2944], [-7.4792, 5.3708]],    # strip 1
    [[316.0517, 360], [-2.4819, 2.5153]]  # strip 2
]
```

## DES-overlap region

```python
overlap_df = []
for (ra_min, ra_max), (dec_min, dec_max) in des_range:
    sel = gz_ds[(gz_ds['RA_deg'] >= ra_min) & (gz_ds['RA_deg'] <= ra_max) &
                (gz_ds['DEC_deg'] >= dec_min) & (gz_ds['DEC_deg'] <= dec_max)]
    overlap_df.append(sel)
overlap_df = pd.concat(overlap_df)
```

```python
plt.hist2d(overlap_df['RA_deg'], overlap_df['DEC_deg'], bins=[200, 100], range=[[0, 360], [-90, 90]], cmap='hot_r')
plt.colorbar(label='Count')
plt.xlabel('RA')
plt.ylabel('Dec')
print(overlap_df.shape[0])
```

```python
gz_el_prob = overlap_df.P_EL_DEBIASED
gz_cs_prob = overlap_df.P_CS_DEBIASED
plt.hist([gz_el_prob, gz_cs_prob], bins=30, range=(0.5, 1), label=['Elliptical', 'Spiral'])
plt.legend()
plt.xlabel('Probability')
plt.ylabel('Galaxy count')
```

```python
def gen_query_for_df(df, query_dir):
    objids_all = df.index
    num_objs = objids_all.size
    bat_size = 5000   # number of objects per query
    if not path.exists(query_dir):
        makedirs(query_dir)

    query_fmt = '''SELECT TOP {bat_size}
       p.objid, p.ra, p.dec, p.u, p.g, p.r, p.i, p.z,
       p.run, p.rerun, p.camcol, p.field,
       p.petroRad_r
    FROM PhotoObj as p
    WHERE 
       p.objid IN ({csl})'''
    for i, start in enumerate(range(0, num_objs, bat_size)):
        end = min(start + bat_size, num_objs)
        bat_ids = objids_all[start:end]
        query_str = query_fmt.format(
            bat_size=bat_ids.size,
            csl=', '.join([str(x) for x in bat_ids])
        )
        with open(path.join(query_dir, 'query-bat%04d.txt' % i), 'w') as fh:
            fh.write(query_str)
```

```python
gen_query_for_df(overlap_df, '../../data/sdss-galaxyzoo/des_overlap/sdss_sql_queries')
```

## High-certainty selection

```python
spiral_thr = sorted(gz_ds['P_CS_DEBIASED'].values)[-20000]
elliptical_thr = sorted(gz_ds['P_EL_DEBIASED'].values)[-20000]
print(spiral_thr, elliptical_thr)
```

```python
certain_df = pd.concat([gz_ds[gz_ds['P_CS_DEBIASED'] >= spiral_thr],
                        gz_ds[gz_ds['P_EL_DEBIASED'] >= elliptical_thr]])
certain_df.shape
```

```python
plt.hist2d(certain_df['RA_deg'], certain_df['DEC_deg'],
           bins=[200, 100], range=[[0, 360], [-90, 90]], cmap='hot_r')
plt.colorbar(label='Count')
plt.xlabel('RA')
plt.ylabel('Dec')
```

```python
gen_query_for_df(certain_df, '../../data/sdss-galaxyzoo/high_certainty/sdss_sql_queries')
```
